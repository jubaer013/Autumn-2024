# 3 Data Visualization with matplotlib and seaborn

**Matplotlib** and **seaborn** are two popular Python libraries for data visualization. Matplotlib provides low-level functionality for creating various types of plots, such as line charts, bar charts, histograms, scatter plots, etc. Seaborn builds on top of matplotlib and offers high-level functions for creating more complex and aesthetically pleasing visualizations, such as heatmaps, boxplots, violin plots, etc. Seaborn also integrates well with pandas, a library for data analysis and manipulation. Together, matplotlib and seaborn can help data scientists and analysts explore and communicate insights from their data.

**3.2 Lineplot**

Drawing a simple line plot of Bangladesh infant mortality rate data

```python
year = ["1993", "1997", "2001", "2004", "2007", "2011", "2014", "2018"]
im_rate = [87, 82, 37, 56, 73, 101, 33, 105]
```

```python
plt.plot(year, im_rate)
plt.show()
```

Adding title and label to the plot

```python
plt.plot(year, im_rate)
plt.title("Bangladesh Infant Mortality Rate")
plt.xlabel("Year")
plt.ylabel("IM")


plt.show()
```

matplotlib allows multiple styles. To use them, you first need to import the style class from matplotlib

```python
from matplotlib import style
```

Run this command to get a list of all possible styles.

```python
print(style.available)
```

```python
# Use style.use(classic) to set a style
style.use("classic")
```

plt.plot() allows few other options. Let's explore them. Try one of these as linestyle {'-', '--', '-.', ':', ''}. Try these as markers ('o', 'v', '^', '<', '>', '8', 's', 'p', '*', 'h', 'H', 'D', 'd', 'P', 'X').

```python
plt.plot(year, im_rate, marker = '^',markerfacecolor='white', color='black', linestyle=':', linewidth=5, markersize=10)
plt.show()
```

Let's compare this with another data. Under 5 mortality rate

```python
u5m_rate = [133, 116, 94, 88, 65, 53, 46, 45]
```

```python
plt.plot(year, im_rate)
plt.plot(year, u5m_rate)

# Add Title and labels
# Then add legend by calling plt.legend()
plt.plot(year, im_rate)
plt.title("Bangladesh Infant Mortality Rate")
plt.xlabel("Year")
plt.ylabel("Infant Mortality")

plt.legend(["Infant","U-5 Infant"])
plt.show()

```

**3.4 Subplot**

Subplots allow us to show multiple plots combined in one plot

```python
plt.subplots_adjust(wspace=0.3, hspace=0.3)
plt.subplot(1, 2, 1)
plt.plot(year, im_rate)
plt.xticks(rotation=90)
plt.subplot(1, 2, 2)
plt.plot(year, u5m_rate)
plt.xticks(rotation=90)
plt.show()
```

**3.5 Contours**

```python
# Manually create the grid data
x = [i for i in range(-5, 6)]
y = [i for i in range(-5, 6)]

# Manually create Z values as a list of lists
Z = []
for i in x:
    row = []
    for j in y:
        z = (i**2 + j**2)**0.5  # Example function
        row.append(z)
    Z.append(row)

# Create the contour plot
plt.contour(x, y, Z)

# Add a title and labels
plt.title('Contour Plot')
plt.xlabel('X axis')
plt.ylabel('Y axis')

# Show the plot
plt.show()
```
